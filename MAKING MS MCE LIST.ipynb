```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('data.csv')
```

```python
ROLL_NUMBER_LIST = {'2k17/mc/'+ str(i) : False for i in range(1,127)}
# print(ROLL_NUMBER_LIST)
```

```python
# names in CSV file for the corresponding coloumn 

_partner1 = 'ROLL NUMBER - STUDENT 1'
_partner2 = 'ROLL NUMBER - STUDENT 2'

def remove_anamoly(param):
    ret_val = []
    for student in df[param]:
        if student is not None and type(student) == str and student[8] == '0':
            student = student[:8] + student[9:]
            student = student.replace(' ','')
            student = student.replace('DTU/','')
            student = student.lower()
            student = student.replace('e','')
        ret_val.append(student)
    return ret_val

df[_partner1] = remove_anamoly(_partner1)
df[_partner2] = remove_anamoly(_partner2)

for student in df[_partner1]:
    print(student)
    if type(student) == str and student is not None and student != '':
        ROLL_NUMBER_LIST[student] = True

for student in df[_partner2]:
    print(student)
    if type(student) == str and student is not None and student != '':
        ROLL_NUMBER_LIST[student.lower()] = True
```
